#

### Nombre total de vÃ©hicules en vente selon le type de vÃ©hicule.

```python
# On regoupe les vÃ©hicules par type en comptant le nombre (de lignes associÃ©es)
# le reset_indesx() permet de transformer la sortie en DataFrame pouur plus de flÃ©xibilitÃ©
nb = data.groupby('vehicleType').size().rename("Nb total de vÃ©hicules").sort_values(ascending=False).reset_index()
nb
```

```python
# Seaborn pour crÃ©er un diagramme en barres
sns.barplot(x='vehicleType', y='Nb total de vÃ©hicules', data=nb)
plt.xticks(rotation = 45) # permert d'Ã©crire les noms de y en diagonale(inclinaison 45Â°)
plt.title('Nombre total de vÃ©hicules en vente selon le type de vÃ©hicule')
plt.xlabel('Type vÃ©hicule')
plt.ylabel('Nombre vÃ©hicule')
plt.show()
```

#

### RÃ©partition des vÃ©hicules en fonction de l'annÃ©e d'immatriculation

```python
nb2 = data.groupby(['yearOfRegistration']).size().rename("Nb total de vÃ©hicules").sort_values(ascending=False).reset_index()
nb2
```

```python
# Seaborn pour crÃ©er un diagramme en barres
filtered_data = nb2[nb2['yearOfRegistration'] > 1980] #car il ya bcp de donnÃ©es == 1 Ã  gauche
plt.figure(figsize=(12, 6))
sns.barplot(x='yearOfRegistration', y='Nb total de vÃ©hicules', data=filtered_data, color='orange')
plt.xticks(rotation = 45)
plt.title("Nombre total de vÃ©hicules en vente selon l'annÃ©e d'immat")
plt.xlabel('AnnÃ©e Immat')
plt.ylabel('Nombre vÃ©hicule')
plt.show()
```

#

### Nombre de vÃ©hicules par marque

```python
nb_vehic_marque = data.groupby('brand').size().rename("Nb vehicule par marque").sort_values(ascending=False).reset_index()
nb_vehic_marque.head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='brand', y='Nb vehicule par marque', data=nb_vehic_marque, color='red')
plt.xticks(rotation = 90)
plt.title("Nombre de vÃ©hicules par marque")
plt.xlabel('Model')
plt.ylabel('Nombre vÃ©hicule')
plt.show()
```

#

### Prix moyen des vÃ©hicules par type de vÃ©hicule et type de boÃ®te de vitesses

```python
prix_moyen = data.groupby(['brand', 'gearbox'])['price'].mean().rename("Prix moyen par type et boÃ®te").sort_values(ascending=False).reset_index()
prix_moyen['Prix moyen par type et boÃ®te'] = prix_moyen['Prix moyen par type et boÃ®te'].round(2)
prix_moyen
# CrÃ©er le graphique en barres avec Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='brand', y='Prix moyen par type et boÃ®te', hue='gearbox', data=prix_moyen)
plt.title('Prix moyen par type et boÃ®te')
plt.xlabel('Marque')
plt.ylabel('Prix moyen')
plt.xticks(rotation=90)  # Rotation des Ã©tiquettes de l'axe x pour une meilleure lisibilitÃ©
plt.legend(title='BoÃ®te de vitesses')
plt.tight_layout()
plt.show()
```

##

### Prix moyen des vÃ©hicules selon le type de carburant et le type de boÃ®te de vitesses

```python
prix_moyen_carb = data.groupby(['fuelType', 'gearbox'])['price'].mean().rename("Prix moyen par type de carburant et boÃ®te").sort_values(ascending=False).reset_index()
prix_moyen_carb['Prix moyen par type de carburant et boÃ®te'] = prix_moyen_carb['Prix moyen par type de carburant et boÃ®te'].round(2)

# CrÃ©er le graphique en barres avec Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='fuelType', y='Prix moyen par type de carburant et boÃ®te', hue='gearbox', data=prix_moyen_carb)
plt.title('Prix moyen par type de carburant et boÃ®te')
plt.xlabel('Marque carburant')
plt.ylabel('Prix moyen')
plt.xticks(rotation=90)  # Rotation des Ã©tiquettes de l'axe x pour une meilleure lisibilitÃ©
plt.legend(title='BoÃ®te de vitesses')
plt.tight_layout()
plt.show()
```

### Benchmarketing Ã  Frankfort

```python
# je selectionne les codes postaux associÃ©s Ã  Frankfort:
Frankfort = [15230, 15232, 15234, 15236]
donnees_filtrees = data[data['postalCode'].isin(Frankfort)]

# CrÃ©er une heatmap avec Seaborn pour les moyennes de prix par 'vehicleType' et 'brand'
# vehicleType gearbox model fuelType brand postalCode
plt.figure(figsize=(12, len('brand')*1.5)) # Pour augmenter la taille du figure
heatmap_data = donnees_filtrees.groupby(['brand', 'vehicleType'])['price'].mean().unstack()
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".1f", cbar=True)
plt.title('Prix moyen par vehicleType et brand pour Frankfort')
plt.xlabel('brand')
plt.ylabel('Vehicle Type')
plt.show()
```
