### Chargement du dataset avec Pandas

```python
import pandas as pd
data = pd.read_csv("cleaned_autos.csv")
data.head(15)
```

#

### Nombre total de véhicules en vente selon le type de véhicule.

```python
# On regoupe les véhicules par type en comptant le nombre (de lignes associées)
# le reset_indesx() permet de transformer la sortie en DataFrame pouur plus de fléxibilité
nb = data.groupby('vehicleType').size().rename("Nb total de véhicules").sort_values(ascending=False).reset_index()
nb
```

```python
# Seaborn pour créer un diagramme en barres
sns.barplot(x='vehicleType', y='Nb total de véhicules', data=nb)
plt.xticks(rotation = 45) # permert d'écrire les noms de y en diagonale(inclinaison 45°)
plt.title('Nombre total de véhicules en vente selon le type de véhicule')
plt.xlabel('Type véhicule')
plt.ylabel('Nombre véhicule')
plt.show()
```

#

### Répartition des véhicules en fonction de l'année d'immatriculation

```python
nb2 = data.groupby(['yearOfRegistration']).size().rename("Nb total de véhicules").sort_values(ascending=False).reset_index()
nb2
```

```python
# Seaborn pour créer un diagramme en barres
filtered_data = nb2[nb2['yearOfRegistration'] > 1980] #car il ya bcp de données == 1 à gauche
plt.figure(figsize=(12, 6))
sns.barplot(x='yearOfRegistration', y='Nb total de véhicules', data=filtered_data, color='orange')
plt.xticks(rotation = 45)
plt.title("Nombre total de véhicules en vente selon l'année d'immat")
plt.xlabel('Année Immat')
plt.ylabel('Nombre véhicule')
plt.show()
```

#

### Nombre de véhicules par marque

```python
nb_vehic_marque = data.groupby('brand').size().rename("Nb vehicule par marque").sort_values(ascending=False).reset_index()
nb_vehic_marque.head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='brand', y='Nb vehicule par marque', data=nb_vehic_marque, color='red')
plt.xticks(rotation = 90)
plt.title("Nombre de véhicules par marque")
plt.xlabel('Model')
plt.ylabel('Nombre véhicule')
plt.show()
```

#

### Prix moyen des véhicules par type de véhicule et type de boîte de vitesses

```python
prix_moyen = data.groupby(['brand', 'gearbox'])['price'].mean().rename("Prix moyen par type et boîte").sort_values(ascending=False).reset_index()
prix_moyen['Prix moyen par type et boîte'] = prix_moyen['Prix moyen par type et boîte'].round(2)
prix_moyen
# Créer le graphique en barres avec Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='brand', y='Prix moyen par type et boîte', hue='gearbox', data=prix_moyen)
plt.title('Prix moyen par type et boîte')
plt.xlabel('Marque')
plt.ylabel('Prix moyen')
plt.xticks(rotation=90)  # Rotation des étiquettes de l'axe x pour une meilleure lisibilité
plt.legend(title='Boîte de vitesses')
plt.tight_layout()
plt.show()
```

##

### Prix moyen des véhicules selon le type de carburant et le type de boîte de vitesses

```python
prix_moyen_carb = data.groupby(['fuelType', 'gearbox'])['price'].mean().rename("Prix moyen par type de carburant et boîte").sort_values(ascending=False).reset_index()
prix_moyen_carb['Prix moyen par type de carburant et boîte'] = prix_moyen_carb['Prix moyen par type de carburant et boîte'].round(2)

# Créer le graphique en barres avec Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='fuelType', y='Prix moyen par type de carburant et boîte', hue='gearbox', data=prix_moyen_carb)
plt.title('Prix moyen par type de carburant et boîte')
plt.xlabel('Marque carburant')
plt.ylabel('Prix moyen')
plt.xticks(rotation=90)  # Rotation des étiquettes de l'axe x pour une meilleure lisibilité
plt.legend(title='Boîte de vitesses')
plt.tight_layout()
plt.show()
```

### Benchmarketing à Frankfort

```python
# je selectionne les codes postaux associés à Frankfort:
Frankfort = [15230, 15232, 15234, 15236]
donnees_filtrees = data[data['postalCode'].isin(Frankfort)]

# Créer une heatmap avec Seaborn pour les moyennes de prix par 'vehicleType' et 'brand'
# vehicleType gearbox model fuelType brand postalCode
plt.figure(figsize=(12, len('brand')*1.5)) # Pour augmenter la taille du figure
heatmap_data = donnees_filtrees.groupby(['brand', 'vehicleType'])['price'].mean().unstack()
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".1f", cbar=True)
plt.title('Prix moyen par vehicleType et brand pour Frankfort')
plt.xlabel('brand')
plt.ylabel('Vehicle Type')
plt.show()
```
